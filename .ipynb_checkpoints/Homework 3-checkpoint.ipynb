```python
def open_csvfile(file_name):
    X=open(file_name,"r")
    Xdata=[]
    
    with X:
        reader = csv.reader(X)
        for row in reader:
            #%for e in row:
            Xdata.append(row)
            
    return Xdata

data_list=open_csvfile('Default.csv')

data_df=pd.DataFrame(data_list,columns=data_list[0])
data_df=data_df.drop([0])
data_df

data_df = pd.read_csv('https://d1pqsl2386xqi9.cloudfront.net/notebooks/Default.csv', index_col=0)
data_df.head()
```

4) Train a model to predict if a student defaults on their loan or not using LogisticRegression based soley on the balance and income variables. Plot the decision boundary in the scatter plot you made in 3) and compute the AUC ROC.


```python

X = data_final[['balance', 'income']]
y = data_final.default_Yes
defaultY = data_final[data_final.default_Yes == 1]
defaultN = data_final[data_final.default_No == 1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20)

clf = LogisticRegression(penalty='l1')
clf.fit(X_train,y_train)

plt.figure(figsize=(15,15))
h = 50
x_min, x_max = data_final[['balance']].min() , data_final[['balance']].max() 
y_min, y_max = data_final[['income']].min() , data_final[['income']].max() 
X, Y = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
plt.xlim(X.min(), X.max())
plt.ylim(Y.min(), Y.max())

AXONE = plt.gca()
AXONE.scatter(defaultN['balance'],defaultN['income'],c='b',label = 'Default_No')
AXONE.scatter(defaultY['balance'],defaultY['income'],c='r',label='Default_Yes')
AXONE.legend(prop={'size':10})
c = clf.predict_proba(np.c_[X.ravel(), Y.ravel()])[:, 1]
c = c.reshape(X.shape)
CS = AXONE.contourf(X, Y, c, cmap='bwr', alpha=.4)
CS2 = AXONE.contour(X, Y, c, cmap='bwr', alpha=.4)
plt.clabel(CS2, fmt = '%2.1f', colors = 'k', fontsize=18)

plt.show()


```

```python

y = data_final['default_Yes']
X = data_final[['balance','income']]
scaler = StandardScaler()
X = scaler.fit_transform(X)

def make_auc(X,y,clf_class,**kwargs):
    random_state = np.random.RandomState(0)
    n_samples, n_features = X.shape
    half = int(n_samples/1.2)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    classifier = clf_class(**kwargs)
    probas_ = classifier.fit(X_train, y_train).predict_proba(X_test)
    
    fpr, tpr, thresholds = roc_curve(y_test, probas_[:, 1])
    roc_auc = auc(fpr, tpr)
    print ("Area under the ROC curve : %f" % (roc_auc))
    return fpr, tpr, roc_auc, thresholds

def ROCplot(fpr,tpr,label,roc_auc):
    plt.clf()
    plt.plot(fpr, tpr, '--',label='ROC curve (Area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')
    #plt.xlim([0.0, 1.0])
    #plt.ylim([0.0, 1.0])
    plt.title('Receiver Operating Characteristic')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc="lower right")
    plt.show()
   

FPR,TPR,ROC, thresholds = make_auc(X,y,LogisticRegression,C=0.1,penalty='l1')
ROCplot(FPR,TPR,'LR',ROC)

```
